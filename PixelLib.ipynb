**Instance Segmentation of live stream video using Pixellib**

```python
import cv2
import pixellib
from pixellib.instance import instance_segmentation

segmented_image = instance_segmentation()
segmented_image.load_model("/content/mask_rcnn_coco.h5") 
camera = cv2.VideoCapture(0)

while camera.isOpened():
    res,frame=camera.read()
    result=segment_image.segmentFrame(frame,show_bboxes=True)
    image=result[1]
    cv2.imshow('Image Segmentation',image)

    if cv2.waitKey(0) & 0xFF==ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```
